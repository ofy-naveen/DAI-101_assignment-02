```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib as plt 
%matplotlib inline
import matplotlib.pyplot as plt
df =sns.load_dataset('tips')
df.head()
```

```python
df.info
```

```python
df['sex'].value_counts(),
```

```python
df['smoker'].value_counts()
```

```python
df['day'].value_counts()
```

```python
df['time'].value_counts()
```

```python
df.columns
```

```python
x = df[['tip', 'sex', 'smoker', 'day', 'time', 'size']]
y = df['total_bill']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.25,random_state=10)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le1=LabelEncoder()
le2=LabelEncoder()
le3=LabelEncoder()
```

```python
X_train['sex'] = le1.fit_transform(X_train['sex'])
X_train['smoker'] =le2.fit_transform(X_train['smoker'])
X_train['time'] =le3.fit_transform(X_train['time'])
```

```python
X_train.head()
```

```python
X_test['sex'] = le1.transform(X_test['sex'])
X_test['smoker'] =le2.transform(X_test['smoker'])
X_test['time'] =le3.transform(X_test['time'])
```

```python
X_test.head()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer(transformers =[('onehot',OneHotEncoder(drop='first'),[3])]
                                    ,remainder='passthrough')
```

```python
X_train=ct.fit_transform(X_train)
```

```python
X_test = ct.transform(X_test)
```

```python
from sklearn.svm import SVR
svr = SVR()
```

```python
svr.fit(X_train,Y_train)
```

```python
Y_pred = svr.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
print (r2_score(Y_test,Y_pred))
print(mean_absolute_error(Y_test,Y_pred))
```

```python
#hyperparamerter tunninng
from sklearn.model_selection import GridSearchCV

#defining parameter range
param_grid = {'C':[0.1,1,10,100,1000],
             'gamma': [1,0.1,0.01,0.001,0.0001],
             'kernel':['rbf']}
```

```python
grid = GridSearchCV(estimator=SVR(), param_grid=param_grid, refit=True, cv=5, verbose=3)
grid.fit(X_train,Y_train)
```

```python
grid.best_params_
```

```python
grid_predict=grid.predict(X_test)
```

```python
print (r2_score(Y_test,grid_predict))
print(mean_absolute_error(Y_test,grid_predict))
```
